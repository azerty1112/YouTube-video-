# 🎬 YouTube Shorts Generator - Colab-friendly (fixed)

This notebook is a Colab-adapted version that: 
- Replaces pyttsx3 (which doesn't work well in Colab) with gTTS.
- Renders text with PIL and uses moviepy ImageClip (avoids ImageMagick/TextClip issues).
- Installs system deps (ffmpeg, fonts) and required Python packages.
- Includes a simple Gradio UI to generate a short video from a title.

Run each cell in order.

```python
#@title 🔧 Core classes: Scenario generator (fallback), TTS (gTTS), Video generator (PIL+moviepy)
import os
import tempfile
from datetime import datetime
from gtts import gTTS
from PIL import Image, ImageDraw, ImageFont
import numpy as np
from moviepy.editor import ColorClip, ImageClip, AudioFileClip, concatenate_videoclips

# Simple scenario generator: uses Ollama if available, otherwise fallback
class OllamaScenarioGenerator:
    def __init__(self):
        self.base_url = os.getenv('OLLAMA_URL', 'http://localhost:11434')
        self.available = self._check()
    def _check(self):
        try:
            import requests
            r = requests.get(f'{self.base_url}/api/tags', timeout=3)
            return r.status_code == 200
        except Exception:
            return False
    def generate_scenario(self, title, max_words=300):
        if not self.available:
            return self._fallback(title)
        try:
            import requests
            prompt = f'Write a short engaging Arabic scenario for a YouTube Short titled: {title} (max {max_words} words)
'
            r = requests.post(f'{self.base_url}/api/generate', json={'model':'qwen','prompt':prompt, 'stream':False}, timeout=30)
            if r.status_code == 200:
                return r.json().get('response','').strip() or self._fallback(title)
        except Exception:
            pass
        return self._fallback(title)
    def _fallback(self, title):
        return f'🎬 سيناريو: {title}

مشهد 1: مقدمة قصيرة وجذابة.
مشهد 2: شرح النقاط الأساسية بسرعة.
مشهد 3: خاتمة ودعوة للاشتراك.'

class TextToSpeechEngine:
    def __init__(self, lang='ar'):
        self.lang = lang
    def text_to_speech(self, text, output_path):
        try:
            tts = gTTS(text=text, lang=self.lang)
            tts.save(output_path)
            return True
        except Exception as e:
            print('TTS error:', e)
            return False

def _find_font():
    candidates = [
        '/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf',
        '/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf'
    ]
    for p in candidates:
        if os.path.exists(p):
            return p
    return None

def render_text_image(text, size=(1080,1920), bg_color=(20,25,50), font_size=80, padding=40):
    # Create an image and draw wrapped text
    W, H = size
    img = Image.new('RGB', (W, H), color=bg_color)
    draw = ImageDraw.Draw(img)
    font_path = _find_font()
    try:
        font = ImageFont.truetype(font_path, font_size) if font_path else ImageFont.load_default()
    except Exception:
        font = ImageFont.load_default()
    # simple wrap by chars (sufficient for short Arabic lines)
    max_width = W - 2*padding
    lines = []
    words = text.split()
    cur = ''
    for w in words:
        test = (w + ' ' + cur).strip() if cur else w
        wsize = draw.textbbox((0,0), test, font=font)[2] if hasattr(draw, 'textbbox') else draw.textsize(test, font=font)[0]
        if wsize <= max_width:
            cur = test
        else:
            if cur:
                lines.append(cur)
            cur = w
    if cur:
        lines.append(cur)
    # vertical centering
    line_h = draw.textbbox((0,0), 'A', font=font)[3] if hasattr(draw, 'textbbox') else draw.textsize('A', font=font)[1]
    total_h = line_h * len(lines) + (len(lines)-1)*10
    y = (H - total_h)//2
    for line in lines:
        w = draw.textbbox((0,0), line, font=font)[2] if hasattr(draw, 'textbbox') else draw.textsize(line, font=font)[0]
        x = (W - w)//2
        draw.text((x, y), line, font=font, fill=(255,255,255))
        y += line_h + 10
    return img

class VideoGenerator:
    VIDEO_DIMENSIONS = (1080, 1920)
    VIDEO_FPS = 30
    VIDEO_DURATION = 20
    def __init__(self):
        pass
    def create_video(self, title, scenario, audio_path, output_path, background_color=(20,25,50)):
        try:
            W, H = self.VIDEO_DIMENSIONS
            # background clip
            bg = ColorClip(size=(W,H), color=background_color, duration=self.VIDEO_DURATION)
            # title image (short duration)
            title_img = render_text_image(title, size=(W,H), bg_color=background_color, font_size=120)
            title_clip = ImageClip(np.array(title_img)).set_duration(4)
            # scenario image (rest of video)
            scenario_short = scenario if len(scenario) < 800 else scenario[:800] + '...'
            scen_img = render_text_image(scenario_short, size=(W,H), bg_color=background_color, font_size=64)
            scen_clip = ImageClip(np.array(scen_img)).set_duration(self.VIDEO_DURATION - 4)
            clips = [title_clip, scen_clip]
            video = concatenate_videoclips(clips)
            # attach audio if exists
            if audio_path and os.path.exists(audio_path):
                audio = AudioFileClip(audio_path)
                if audio.duration > self.VIDEO_DURATION:
                    audio = audio.subclip(0, self.VIDEO_DURATION)
                video = video.set_audio(audio)
            video = video.set_fps(self.VIDEO_FPS)
            video.write_videofile(output_path, fps=self.VIDEO_FPS, codec='libx264', audio_codec='aac')
            return True
        except Exception as e:
            print('Video generation error:', e)
            return False

print('Core classes loaded: OllamaScenarioGenerator, TextToSpeechEngine (gTTS), VideoGenerator')
```

```python
#@title ▶️ Simple demo: generate a video from a title
from pathlib import Path
tmp = tempfile.mkdtemp(prefix='ytshorts_')
print('Temp dir:', tmp)
sc = OllamaScenarioGenerator()
tts = TextToSpeechEngine(lang='ar')
vg = VideoGenerator()

def make_short(title, upload_to_github=False):
    scenario = sc.generate_scenario(title)
    audio_path = os.path.join(tmp, 'audio.mp3')
    ok = tts.text_to_speech(scenario, audio_path)
    video_path = os.path.join(tmp, f'{title[:20].replace('/','_')}.mp4')
    ok2 = vg.create_video(title, scenario, audio_path if ok else None, video_path)
    return ok2, video_path, scenario

# Example run (uncomment to test)
# ok, vpath, scen = make_short('كيفية تعلم البرمجة من الصفر')
# print('Done:', ok, vpath)
```

```python
#@title 🌐 Gradio UI (run to serve)
import gradio as gr

def ui_generate(title):
    if not title or len(title.strip()) < 3:
        return 'Please provide a longer title', None, ''
    ok, vpath, scen = make_short(title)
    if ok:
        return '✅ Video created', vpath, scen
    return '❌ Failed to create video', None, ''

with gr.Blocks() as demo:
    gr.Markdown('# 🎬 YouTube Shorts Generator - Colab-friendly')
    title_in = gr.Textbox(label='Video Title')
    out_status = gr.Textbox(interactive=False, label='Status')
    out_video = gr.Video(label='Output video')
    out_scenario = gr.Textbox(interactive=False, label='Scenario')
    btn = gr.Button('Generate')
    btn.click(fn=ui_generate, inputs=[title_in], outputs=[out_status, out_video, out_scenario])

print('Run demo.launch() to start the Gradio UI (or call demo.launch(share=True))')
demo
```
